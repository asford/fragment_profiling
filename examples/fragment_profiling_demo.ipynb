Import and initalize rosetta to parse input poses.

```python
import rosetta; rosetta.init()
from rosetta.core.scoring.dssp import Dssp
```

StructureDatabase class served as interface to backend structure store. FragmentSpecifiction is used to define the length of fragments extracted from source structures.

```python
from interface_fragment_matching.structure_database.store import StructureDatabase
from interface_fragment_matching.fragment_fitting.store import FragmentSpecification
```

`BenchmarkedProfileFragmentQualityTask` worker class handles loading backing residues, preparing data for sequence scoring, and execution of profile runs for input structures.

`BenchmarkedProfileFragmentQualityTask` returns `ProfileFragmentQualityResult` objects, which store query fragments, result fragments, rmsd analysis results, and manage plotting result summaries.

`FragmentProfilerParameters` define the parameters needed to match a query to benchmark data.

```python
from fragment_profiling.tasks import BenchmarkedProfileFragmentQualityTask
from fragment_profiling.profile_fragment_quality import FragmentProfilerParameters
```

```python
profiler_parameters = FragmentProfilerParameters(FragmentSpecification(9, "CA"), "blosum100", 200)
profiler_task = BenchmarkedProfileFragmentQualityTask("/work/fordas/workspace/fragment_profiling/benchmarking/vall_store_fragment_profiling.h5", "vall_benchmarking", profiler_parameters)
#Not needed, but validates configuration values before running the profiler. `setup()` will be run if needed before profiling
profiler_task.setup()
```

Load positive and negative example structures and extract residue entries for profiling.

Perform DSSP secondary structure assignment to identify loop regions.

```python
positive_structure = rosetta.pose_from_pdb("6h_3l_3_recirc_full_min_0006.good_fragments.pdb")
Dssp(positive_structure).insert_ss_into_pose(positive_structure)

negative_structure = rosetta.pose_from_pdb("6h2l_remodel_5.poor_fragments.pdb")
Dssp(negative_structure).insert_ss_into_pose(negative_structure)
```

```python
positive_structure_residues = StructureDatabase.extract_residue_entries_from_pose(positive_structure)
negative_structure_residues = StructureDatabase.extract_residue_entries_from_pose(negative_structure)
```

Call task object with input residues to perform profiling and retrieve results.

```python
positive_structure_profile_results = profiler_task(positive_structure_residues)
negative_structure_profile_results = profiler_task(negative_structure_residues)
```

Profiling can be performed on a subset of structure residues to improve throughput.

Select fragments spanning loop regions in the source structures.

Fragment arrays are shape `N` structured arrays, containing shape `fragment_length` subfields. Selecting subfields results in shape `(N, fragment_length)` sub-arrays.

Count the number of loop (`L`) secondary structure assignments per fragment, and select fragments with 1 or more `L` residues.

```python
positive_structure_fragment_indicies, positive_structure_fragments = profiler_parameters.fragment_specification.fragments_from_source_residues(positive_structure_residues, additional_per_residue_fields=["ss", "sc"])
positive_loop_fragments = positive_structure_fragments[numpy.sum( positive_structure_fragments["ss"] == "L", axis=-1) != 0]

negative_structure_fragment_indicies, negative_structure_fragments = profiler_parameters.fragment_specification.fragments_from_source_residues(negative_structure_residues, additional_per_residue_fields=["ss", "sc"])
negative_loop_fragments = negative_structure_fragments[numpy.sum( negative_structure_fragments["ss"] == "L", axis=-1) != 0]
```

```python
positive_loop_structure_profile_results = profiler_task.profiler.perform_fragment_analysis(positive_loop_fragments)
negative_loop_structure_profile_results = profiler_task.profiler.perform_fragment_analysis(negative_loop_fragments)
```

```python
f, (ax1, ax2) = pylab.subplots(nrows=2)

negative_loop_structure_profile_results.plot_all_fragment_analysis(ax1)
ax1.set_title("negative_loop_structure_profile_results: " + ax1.title.get_text())

positive_loop_structure_profile_results.plot_all_fragment_analysis(ax2)
ax2.set_title("positive_loop_structure_profile_results: " + ax2.title.get_text())

f.tight_layout()
pass
```
